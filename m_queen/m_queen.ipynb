```python
from datetime import datetime as time
```

```python
def promising(board, row, col, n):
    for i in range(col):
        if board[row][i] == 1:
            return False
    for i, j in zip(range(row, -1, -1), range(col, -1, -1)):
        if board[i][j] == 1:
            return False
    for i, j in zip(range(row, n, 1), range(col, -1, -1)):
        if board[i][j] == 1:
            return False
    return True

def solve_queen(board, col, n, solutions):
    if col == n:
        solutions.append([board[i].index(1) + 1 for i in range(n)])
        return
    for i in range(n):
        if promising(board, i, col, n):
            board[i][col] = 1
            solve_queen(board, col + 1, n, solutions)
            board[i][col] = 0

def m_queen(m):
    board = [[0 for _ in range(m)] for _ in range(m)]
    solutions = []
    solve_queen(board, 0, m, solutions)
    return solutions

```

# Performance on i5 CPU (8 MB CACH) and 12 GB of RAM

```python
time_list = []
answers = {}
for _ in range(4, 21):
    start_time = time.now()
    answers[_] = m_queen(_)    
    last_time = time.now()
    taken_time = last_time-start_time
    
    time_list.append(taken_time)
    print(f"performance of the algorithm on {_}*{_} and for {_}_queens : {answers[_]} \n the taken time to run the algorithm for {_}_queens: {taken_time} \n \n \n \n")
```
